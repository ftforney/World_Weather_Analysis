```python
#Import everything we will be using.

import numpy as np
import timeit
import pandas as pd
import matplotlib.pyplot as plt
from citipy import citipy
import requests
from config import weather_api_key
import time
from datetime import datetime
```

```python
#Defining a URL we will use to grab information from a website.

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
#Generate 2,000 random latitudes and longitudes

lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
#Converting our random generation to a list of lats and longs

coordinates = list(lat_lngs)
```

```python
#Validating the data and making sure citipy is working...

for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
#Initialize a list for our cities and respective countries...

cities = {'city':[],'country':[]}
# Get nearest city for each coordinate...

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    country = citipy.nearest_city(coordinate[0], coordinate[1]).country_code
    # Make sure there are no duplicates...
    
    if city not in cities['city']:
        cities['city'].append(city)
        cities['country'].append(country)
# Test that the process worked and print how many cities we got...
print(cities['city'][730],cities['country'][730])
len(cities['country'])
```

```python
#Start the process of gathering the data...

city_data = []
print("Test Request                 ")
print("-----------------------------")
city_url = url + "&q=" + cities['city'][2].replace(" ","+") + "%2C" + cities['country'][2]
city_weather = requests.get(city_url).json()
print(city_weather)
print("-----------------------------")
print("Beginning Data Retrieval     ")
print("-----------------------------")

record_count = 1
set_count = 1
city_count = 0
```

```python
for i, city in enumerate(cities['city']):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(60)

    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+") + "%2C" + cities['country'][city_count]

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}, {cities['country'][city_count]}")
    # Add 1 to the record count.
    record_count += 1
    # Increment the city count so we can get the correct country code...
    city_count += 1
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_weather_desc = city_weather["weather"][0]["description"]
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City":city.title(),
                          "Lat":city_lat,
                          "Lng":city_lng,
                          "Max Temp":city_max_temp,
                          "Humidity":city_humidity,
                          "Cloudiness":city_clouds,
                          "Wind Speed":city_wind,
                          "Country":city_country,
                          "Date":city_date,
                          "Current Description":city_weather_desc})
        print(f"City count is now {city_count}")

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        print(f"City count is now {city_count}")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
city_count = 0
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Create the output file.
output_data_file = "WeatherPy_Database.csv"
# Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
